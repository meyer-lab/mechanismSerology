```python
from tensordata.zohar import data3D as zohar

from maserol.core import optimizeLoss
from maserol.preprocess import prepare_data
```

```python
from maserol.preprocess import HIgGs


data = prepare_data(zohar(xarray=True, logscale=False))
opt_kwargs = {
    "metric": "mean",
    "fitKa": False,
    "ab_types": HIgGs,
}
x, fun, loss_traj = optimizeLoss(data, **opt_kwargs)
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(loss_traj.shape[0])
plt.plot(x, loss_traj)
print(fun)
```
